```python
array = ["水","金","地","火","木","土","天","海","冥"]

# for文
for i, planet in enumerate(array, start=1):
    print(f"{i}{planet}")

# while文
num = 10
index = 0

while index < len(array):
    print(f"{num}{array[index]}")
    num += 1
    index += 1
```
